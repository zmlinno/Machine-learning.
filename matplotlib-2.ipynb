### 도표

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi,50)
plt.figure(figsize=(9,6))
y1 = np.sin(x)
y2 = np.cos(x)
# plt.plot(x,y1,label = 'Sin')
# plt.plot(x,y2,label = 'Cos')
plt.plot(x,y1)
plt.plot(x,y2)
# (x, y, width, height)
# plt.legend(loc = 'upper right',
#            fontsize = 20,
#            ncol = 2,bbox_to_anchor = [0,0,0.02,0.04]) # 显示图例
plt.legend(labels = ['Sin','Cos'],loc = 'center',
           fontsize = 20,
           ncol = 2,bbox_to_anchor = [0.5,1.1,0,0]) # 显示图例
```

### 척추 이동

```python
plt.figure(figsize=(9,6))

plt.plot(x,y1,x,y2)

ax = plt.gca() # 현재 시각화된 Axes를 가져옵니다，axis[axis = 1,0]

# 우측과 상단의 척추를 없앱니다
ax.spines['right'].set_color('white')
ax.spines['top'].set_alpha(0)

# 이동합니다.
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',np.pi))
# plt.yticks은 plt를 기본으로 하기 때문에 현재 시각화된 것을 설정합니다
# _ = plt.yticks([-1,0,1]) # plt, 최상위 변수, import matplotlib.pyplot as pltimport matplotlib.pyplot as plt
ax.set_yticks([-1,0,1]) #  ax는 현재 시각화된 것을 나타냅니다
```

```python
bottom = ax.spines['bottom']
bottom.set_position()
```

### 도표저장

```python
# fig 그림, 그림은 시각화된 것을 포함하고 여러 개를 포함할 수 있습니다.
fig = plt.figure(linewidth = 16,facecolor='m',edgecolor='lightgreen') # 효과를 볼 수 없습니다.
# plt.figure(figsize=(9,6))
fig.set_figwidth(9)
fig.set_figheight(6)
plt.plot(x,y1,x,y2)

ax = plt.gca() # 뷰
ax.set_facecolor('lightgreen')

plt.legend(['Sin','Cos'],loc = 'center',bbox_to_anchor = [0.5,1.1,0,0],ncol = 2,fontsize = 20)
plt.title('Sin Vs Cos',pad = 60)
plt.ylabel('f(x) = sin(x) vs cos(x)',rotation = 0,fontsize = 20,labelpad = 100)
plt.savefig('./2.pdf',dpi = 300,bbox_inches = 'tight')
```

```python
import numpy as np
import pandas as pd
def f(x):
    return np.exp(-x) * np.cos(2*np.pi*x)
x = np.linspace(0,5,50)
plt.figure(figsize=(9,6))
plt.plot(x,f(x),color = 'purple',
         marker = 'o',
         ls = '--',
         lw = 2,
         alpha = 0.6,
         markerfacecolor = 'red',# 마커 색상
         markersize = 10,# 마커 크기
         markeredgecolor = 'green',#마커 테두리 색상
         markeredgewidth = 2)#마커 테두리 너비
plt.xticks(size = 18) # 눈금 크기 설정
_ = plt.yticks(size = 18)
```

### 중첩

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi,np.pi,25)
y = np.sin(x)
fig = plt.figure(figsize=(9,6)) 

plt.plot(x,y)


ax = plt.axes([0.2,0.55,0.3,0.3],anchor ='NE' ) 
ax.plot(x,y,color = 'g')


ax = fig.add_axes([0.55,0.2,0.3,0.3]) 
ax.plot(x,y,color = 'r')
```

### 이중 축 레이아웃

```python
x = np.linspace(-np.pi,np.pi,100)
y1 = np.sin(x)
y2 = np.exp(x)

plt.figure(figsize=(9,6))
plt.rcParams['font.size'] = 20

ax1 = plt.gca()
ax1.plot(x,y2,color = 'green')
ax1.set_xlabel('time(s)')
ax1.set_ylabel('exp',color =  'green')

ax2 = ax1.twinx() 
ax2.plot(x,y1,color = 'red')
ax2.set_ylabel('sin',color = 'red')
plt.tight_layout() 
plt.savefig('./双轴布局.png',dpi = 100)
```
