#### map字典映射

```python
# df['Math'] 取出了一列，这一列是【Series】
# map函数，是Series的函数
df['Math'].map({47:88,8:108,32:66})
```

```python
df['Math'].replace({47:88,8:108,32:66})
```

```python
# 函数名任意
# 功能自己定义
# 非常强大
def convert(x):# x表示什么，代表Math这一列的每一个元素【数值】
    if x == 47:
        return 88
    elif x == 8:
        return 108
    elif x == 32:
        return 66
    else:
        return x
# 原数据就发生了改变
df['Math'] = df['Math'].map(convert)
```

```python
df['Python'].map(lambda x : 88 if x == 60 else 149)
```

#### 随机抽样重排哑变量【独热编码】

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(data = np.random.randint(0,10,size = (10,3)),
                  index = list('ABCDEFHIJK'),
                  columns=['Python','Tensorflow','Keras'])
display(df)
index = np.random.permutation(10) # 随机重排0~9数字重排
print(index)
df.take(index) # 重排DataFrame
```

```python
df.take(np.random.randint(0,10,size = 30)) # 随机抽样
```

```python
# 哑变量，独热编码，1表示有，0表示没有
df = pd.DataFrame({'key':['b','b','a','c','a','b']})
display(df)
```

```python
pd.get_dummies(df,prefix='',prefix_sep='')
```

### 分箱操作

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(data = np.random.randint(0,150,size = (50,3)),
                  columns=['Python','Tensorflow','Keras'])
display(df)
# 1、等宽分箱
pd.cut(df.Python,bins = 3,labels=['不及格','中等','优秀']) # 参数bins = 3表示将Python这一列分成3份
# 随机数范围0到150，样本比较多，所以最小值和最大值：0,149
# 0到50,51到100,101到149
```

```python
# 指定宽度分箱
pd.cut(df['Keras'],#分箱数据 # df.Keras
       bins = [0,60,90,120,150],#分箱断点
       right = False,# 左闭右开
       labels=['不及格','中等','良好','优秀'])# 分箱后分类
```

```python
# 2、等频分箱
result = pd.qcut(df.Python,q = 4,# 4等分
        labels=['差','中','良','优']) # 分箱后分类
result
```

```python
result.value_counts()
```

```python
pd.cut(df.Python,bins = 4,labels=['差','中','良','优']).value_counts()
```
